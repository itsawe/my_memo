# 1. PandasDummy Variables

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
dummies = pd.get_dummies(df['town'])
dummies
```

```python
merged = pd.concat([df,dummies], axis = 'columns')
merged
```

```python
final = merged.drop(['town', 'west windsor'], axis = 'columns')
final
```

```python
model = linear_model.LinearRegression()
```

```python
X = final.drop(['price'], axis = 'columns')
X
```

```python
y = final.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
#if we want to check how good is the model, we can use score method
model.score(X,y)
```

# Exercise

```python
carprices = pd.read_csv('carprices.csv')
carprices
```

```python
cp = carprices.rename({"Sell Price($)" : "sell_price" , "Age(yrs)" : "age", "Car Model" : "car_model", "Mileage" : "mileage"}, axis='columns')
cp
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.scatter(cp.mileage, cp.sell_price, c='red',marker='x')
plt.xlabel('Car Mileage')
plt.ylabel('Price ($)')
```

```python
dummies_cp = pd.get_dummies(cp.car_model)
dummies_cp
```

```python
merged_cp = pd.concat([cp,dummies_cp], axis='columns')
merged_cp
```

```python
final_cp = merged_cp.drop(['car_model', 'Mercedez Benz C class'], axis = 1)

final_cp
```

```python
X_cp = final_cp.drop(['sell_price'], axis=1)
X_cp
```

```python
y_cp = final_cp.sell_price
y_cp
```

```python
model_cp = linear_model.LinearRegression()

model_cp.fit(X_cp,y_cp)
```

```python
model_cp.score(X_cp,y_cp)
```

__Price of mercedez benz that is 4 yr old with mileage 45000__

```python
model_cp.predict([[45000, 4,0,0]])
```

```python
model_cp.predict([[86000,7,0,1]])
```
